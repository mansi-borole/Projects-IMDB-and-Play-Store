```python
play['Rating'] = pd.to_numeric(play['Rating'], errors='coerce')
play['Rating'].dtype
```

```python
play['Size'].unique()
```

```python
play['Size'] = play.Size.replace("Varies with device",np.nan)
play['Size'] = play.Size.str.replace("M","000") # All size values became the kilobyte type.
play['Size'] = play.Size.str.replace("k","")
play['Size'] = play.Size.replace("1,000+",1000)
play['Size'] =play['Size'].astype(float)
play['Size'].dtype
```

```python
play['Installs'] = play.Installs.str.replace(",","")
play['Installs'] = play.Installs.str.replace("+","")
play['Installs'] = play.Installs.replace("Free",np.nan)
play['Installs'] = play['Installs'].astype(float)
play['Installs'].dtype
```

```python
play.shape
```

```python
play[play["Reviews"].str.isdigit()==False]
```

```python
def clean_reviews(x):
    if(x.isdigit()):
        return int(x)
    else:
        return float(x[:-1]) * 1000000
```

```python
play["Reviews_"] = play["Reviews"].apply(clean_reviews)
```

```python
play["Size"]
```

```python
def clean_size(x):
    if x.isdigit():
        return float(x)
    elif(x[-1]=="M"):
        return float(x[:-1]) * 1e06
    elif(x[-1]=="k"):
        return float(x[:-1]) * 1e03
    else:
        return np.nan
    
```

```python
play["Size_"] = play["Size"].apply(clean_size)
```

```python
play["Price"].value_counts()
```

```python
def clean_price(x):
    if x[0]=="$":
        return float(x[1:])
    else:
        return 0
```

```python
play["Price_"] = play["Price"].apply(lambda x: float(x[1:]) if x[0]=='$' else 0)
```

```python
play["Price_"].value_counts()
```

```python
play["Genres"]
```

```python
play["Genre1"] = play["Genres"].apply(lambda x:x.split(";")[0])
```

```python
play["Genre2"] = play["Genres"].apply(lambda x:x.split(";")[1] if ";" in x else x)
```

```python
def clean_lastupdated(x):
    if x=="1.0.19":
        return np.nan
    else:
        return pd.to_datetime(x)
```

```python
play["Last Updated_"]= play["Last Updated"].apply(clean_lastupdated)
```
